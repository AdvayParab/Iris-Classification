# Model Training and Testing

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
# Split the data into input and output

X = df.drop(columns=['Species'])

Y = df['Species']

# Split the data for training and testing

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)

# Train the Logistic Regression model

modellr = LogisticRegression()
modellr.fit(x_train, y_train)

# Evaluate the model on the testing data
print("Logistic Regression Accuracy: ", modellr.score(x_test, y_test) * 100)
```

```python
modellr.fit(x_train.values,y_train.values)
```

```python
print("Accuracy:",modellr.score(x_test,y_test)*100)
```

```python
# Train the K-nearest neighbors model
modelkn = KNeighborsClassifier()
modelkn.fit(x_train.values, y_train.values)

# Evaluate the model on the testing data
print("K- nearest neighbors Accuracy: ", modelkn.score(x_test, y_test) * 100)
```

```python
modelkn.fit(x_train.values, y_train.values)
```

```python
print("Accuracy: ", modelkn.score(x_test, y_test) * 100)
```

```python
# Decision Tree 
modeldt = DecisionTreeClassifier()
modeldt.fit(x_train.values,y_train.values)
print("Decision Tree Accuracy: ", modeldt.score(x_test, y_test) * 100)
```
